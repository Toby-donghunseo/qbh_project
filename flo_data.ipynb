```python
data120k_dict = data120k.to_dict('records')
for x in data120k_dict:
    if not isinstance(x['genre_id_basket'], str):
        x['genre_id_basket'] = []
    else:
        x['genre_id_basket'] =[ast.literal_eval(y) for y in x['genre_id_basket'] .split(',')]
    if not isinstance(x['genre_name_basket'], str):
        x['genre_name_basket'] = []
    else:
        x['genre_name_basket'] = x['genre_name_basket'].split(',')
    if not isinstance(x['artist_id'], str):
        x['artist_id'] = -1
    else:
        x['artist_id'] = ast.literal_eval(x['artist_id'])
    if not isinstance(x['artist_name_basket'], str):
        x['artist_name_basket'] = []
    else:
        x['artist_name_basket'] = x['artist_name_basket'].split(',')
```
